```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageEnhance, ImageOps
from vazio.utils import walk
```

```python
def plot(img_dict1, img_dict2):
    plt.gray()
    fig=plt.figure(figsize=(16, 16))
    i = 1
    for (name1, img1), (name2, img2) in zip(img_dict1.items(), img_dict2.items()):
        ax = fig.add_subplot(8, 4, i)
        ax.set_title(name1)
        plt.imshow(img1)
        i+=1
        ax = fig.add_subplot(8, 4, i)
        ax.set_title(name2)
        plt.imshow(img2)
        i+=1
    plt.show()
```

```python
def load_apply_function(images, function, params={}):
    loaded_imgs = {}
    enhanced_imgs = {}
    for filename in images[:16]:
        name = filename.split('/')[-1]
        pil_image = Image.open(filename)
        loaded_imgs[name] = pil_image
        enhanced_imgs[name] = function(pil_image, **params)
    return loaded_imgs, enhanced_imgs
```

```python
images = walk('vazio/naovazios')
```

```python
loaded_imgs = {}
enhanced_imgs = {}
for filename in images[:16]:
    name = filename.split('/')[-1]
    pil_image = Image.open(filename)
    loaded_imgs[name] = pil_image
    enhanced_imgs[name] = ImageOps.equalize(pil_image)
```

```python
loaded_imgs, enhanced_imgs = load_apply_function(images, ImageOps.autocontrast, {'cutoff': 10})
```

```python
plot(loaded_imgs, enhanced_imgs)
```
