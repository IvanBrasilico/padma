```python
import pickle
list = pickle.dumps(X[0])

print(len(list))
decode = pickle.loads(list)
print(X[0] == decode)
print(list)
```

```python
import sqlite3 as sql
import time

con = sql.connect('test.db',isolation_level=None)
cur = con.cursor()
cur.execute("DROP TABLE IF EXISTS numpytest")
con.commit()
cur.execute("CREATE TABLE numpytest (id INTEGER PRIMARY KEY, array BLOB)")
con.commit()
```

```python
t0 = time.time()
cur.execute("BEGIN")
for r in range(0, X.shape[0]):
    cur.execute("INSERT INTO numpytest VALUES (?,?)", (None, pickle.dumps(X[r])) )
cur.execute("END")
con.commit()
t1 = time.time()
print("% segundos decorridos...", (t1-t0))
print("% segundo por registro.", (t1-t0)/X.shape[0])
X
```

```python
t0 = time.time()
cur.execute("SELECT array FROM numpytest")
data = cur.fetchall()
t1 = time.time()
X2 = np.empty((len(data), 200))
for r in range(0, len(data)):
    Y = np.array(pickle.loads(data[r][0]), dtype=np.float32)
    X2 [r] = Y
t2 = time.time()
print("% segundos decorridos para o FETCH", (t1-t0))
print("% segundos para conversão e montagem do array.", (t2-t1))
X2
```
